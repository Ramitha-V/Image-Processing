```python
city_block=abs(P[0]-Q[0])+abs(P[1]-Q[1]) #Manhattan distance
city_block
```

```python
chess_board=max(abs(P[0]-Q[0]),abs(P[1]-Q[1])) #Chebyshev distance
chess_board
```

```python
from scipy.spatial import distance
ed=distance.euclidean(P,Q)
cb=distance.cityblock(P,Q) #Manhattan
chb=distance.chebyshev(P,Q) #Chess-board
```
