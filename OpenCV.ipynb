```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

#### Reading the Image

```python
image = cv2.imread(r'D:\SEM 3\Fundamentals of Image Processing Lab\gojo.jpg')
```

#### Showing the Image

```python
plt.imshow(image)
plt.show()
```

```python
bgr_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(bgr_image)
plt.show()
```

#### Converting to Gray Scale

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.show()
```

#### Resizing Image

```python
# Define new dimensions for resizing
new_height = 300
new_width = 200

# Resize the image
resized_image = cv2.resize(image, (new_width, new_height))

# Display the resized image
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.show()
```

#### Blurring Image

```python
blurred_image = cv2.GaussianBlur(image, (9, 9), 0)

# Display the blurred image
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.show()
```

#### Splitting into individual channels

```python
# Split the BGR image into its individual channels
blue_channel, green_channel, red_channel = cv2.split(bgr_image)

# Display each channel separately
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel')

plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='gray')
plt.title('Green Channel')

plt.subplot(1, 3, 3)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')

plt.show()
```
