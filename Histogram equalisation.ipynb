```python
plt.suptitle("Case 1:Left Concentrated Equalisation")
# Low contrast: left concentrated
# Subplot 1: Original Image
plt.subplot(2, 2, 1)
plt.title("Image before equalization")
plt.imshow(img_gray,cmap='gray')

# Subplot 2: Histogram before equalization
plt.subplot(2, 2, 2)
plt.title("Histogram before equalization")
img_hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(img_hist)


# Perform histogram equalization
equalised_hist = cv2.equalizeHist(img_gray)

plt.subplot(2, 2, 3)
plt.title("Image after equalization")
plt.imshow(equalised_hist, cmap='gray')


plt.subplot(2,2,4)
plt.title("Histogram after equalization")
img_hist_equalised = cv2.calcHist([equalised_hist], [0], None, [256], [0, 256])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(img_hist_equalised)

plt.tight_layout()
plt.show()

```

```python
plt.suptitle("Case 2:Right Concentrated Equalisation")
# Low contrast: Right concentrated
# Subplot 1: Original Image
plt.subplot(2, 2, 1)
plt.title("Image before equalization")
plt.imshow(img1_gray,cmap='gray')

# Subplot 2: Histogram before equalization
plt.subplot(2, 2, 2)
plt.title("Histogram before equalization")
img1_hist = cv2.calcHist([img_1], [0], None, [256], [0, 256])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(img1_hist)


# Perform histogram equalization
equalised_hist1 = cv2.equalizeHist(img1_gray)

plt.subplot(2, 2, 3)
plt.title("Image after equalization")
plt.imshow(equalised_hist1, cmap='gray')


plt.subplot(2,2,4)
plt.title("Histogram after equalization")
img_hist_equalised1 = cv2.calcHist([equalised_hist1], [0], None, [256], [0, 256])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(img_hist_equalised1)

plt.tight_layout()
plt.show()

```
