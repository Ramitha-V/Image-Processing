```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('D:\SEM 3\Flexi Credit Course-2\gojo.jpg')
```

```python
img
```

```python
plt.imshow(img)
plt.show()
```

```python
#Converting RGB into BGR
img_new=cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img_new)
plt.show()
```

```python
img_new.shape # BGR in openCV #rows-height #columns-width
```

```python
width=img_new.shape[1]
width
```

```python
height=img_new.shape[0]
height
```

```python
print("The width of the original image is",width,", the height of the original image is",height)
```

```python
# Resize the image
W=1024
H=1024

# increase in image: interpolation=cv2.INTER_NEAREST 
# decrease in image: interpolation=cv2.INTER_AREA

new=cv2.resize(img_new,(W,H),interpolation=cv2.INTER_NEAREST)
```

```python
plt.imshow(new)
plt.show()
```

```python
plt.imshow(img_new)
plt.show()
```

```python
# Aspect ratio
# scale 100= keeps the same image and scale<100:downscale and scale>100:upscale
scale=50 # exactly half
W1=int(img_new.shape[1]*scale/100)
H1=int(img_new.shape[0]*scale/100)

new_img=cv2.resize(img_new,(W1,H1),interpolation=cv2.INTER_AREA)
```

```python
plt.imshow(new_img)
plt.show()

print(new_img.shape)
```

```python
# only three options 90 clockwise, 90 anticlockwise 180
r_img=cv2.rotate(img_new,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(r_img)
plt.show()
```

```python
# only three options 90 clockwise, 90 anticlockwise 180
r_img=cv2.rotate(img_new,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(r_img)
plt.show()
```

```python
# only three options 90 clockwise, 90 anticlockwise 180
r_img=cv2.rotate(img_new,cv2.ROTATE_180)
plt.imshow(r_img)
plt.show()
```

```python
width=img_new.shape[1]
height=img_new.shape[0]

center=(width//2,height//2)

rotate_matrix=cv2.getRotationMatrix2D(center=center,angle=45,scale=1)
print(rotate_matrix)

rotated_image=cv2.warpAffine(src=img_new,M=rotate_matrix,dsize=(width,height))

plt.imshow(rotated_image)
plt.show()
```
